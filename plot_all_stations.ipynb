```python
from bokeh.io import output_file, show, output_server
from bokeh.models import (GMapPlot, GMapOptions, ColumnDataSource, Circle, DataRange1d, PanTool, WheelZoomTool, BoxSelectTool)
import pickle
from bokeh.io import output_notebook
import matplotlib as mpl
import pylab as pl
from bokeh.models import HoverTool, ResetTool
```

Now the pickle file that stores the latitude and longitude of all the stations is imported

```python
lat_lon_df = pickle.load(open("data/dep_times_lat_lon.p","rb"))
col_vals = lat_lon_df['num_deps'].values

colors = [
    "#%02x%02x%02x" % (int(r), int(g), int(b)) for r, g, b, _ in 255*pl.cm.Blues(mpl.colors.Normalize()(col_vals))
]
lat_lon_df['col_vals'] = colors
```

```python
#output_server("Test server")
output_notebook()
```

Give the specifications of the map to be displayed

```python
map_options = GMapOptions(lat=23.29, lng=78.73,map_type="roadmap", zoom=4)
```

Initiate the map

```python
plot = GMapPlot(
    x_range=DataRange1d(), y_range=DataRange1d(), map_options=map_options, title="India"
)
```

```python
source = ColumnDataSource(lat_lon_df)

circle = Circle(x ='lon', y = 'lat',size = 5, fill_color = 'col_vals')
plot.add_glyph(source, circle)
```

```python
plot.add_tools(PanTool(), WheelZoomTool(), BoxSelectTool(), HoverTool(tooltips = [('name','@name'),('No. of deps/week', '@num_deps'),('Code','@code')]),ResetTool())
```

```python
show(plot)
```
